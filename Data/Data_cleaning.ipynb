```python
import pandas as pd

# Read the CSV file into a DataFrame
coffee_df = pd.read_csv('coffee_fix.csv')

# Display the first few rows of the DataFrame to verify the content
coffee_df.head()
```

```python
coffee_df.isnull().sum()
```

```python
# Drop the columns in place
coffee_df.drop(columns=['all_text', 'agtron'], inplace=True)

# Display the updated DataFrame
coffee_df.head()
```

```python
import pandas as pd
import re  # Import regular expressions module


# Define conversion rates (example rates)
conversion_rates = {
    'NT': 0.032,  # New Taiwan Dollar to USD (example rate)
    'NT$': 0.032,
}

def clean_price_string(price_str):
    # Correct malformed numbers like '19..75' to '19.75'
    price_str = re.sub(r'(\d+)\.\.(\d+)', r'\1.\2', price_str)
    return price_str

def convert_to_cost_per_12_ounces(price_str):
    # Convert price_str to a string to avoid errors
    if not isinstance(price_str, str):
        return None
    
    # Clean the price string
    price_str = clean_price_string(price_str)
    
    # Extract the currency, amount, and weight from the price string
    match = re.match(r'([A-Za-z ]*)\$?([\d,.]+)[ /]*(\d+)\s*(ounces|grams)?', price_str)
    
    if match:
        currency = match.group(1).strip()  # Currency symbol or name
        amount = float(match.group(2).replace(',', ''))  # Price amount, remove commas
        weight = float(match.group(3))  # Weight
        unit = match.group(4)  # Unit of weight
        
        # Skip if weight is zero or missing
        if weight == 0:
            return None
        
        # Convert currency to USD if necessary
        if currency in conversion_rates:
            amount *= conversion_rates[currency]
        
        # Convert grams to ounces if necessary
        if unit == 'grams':
            weight /= 28.3495  # 1 ounce = 28.3495 grams
        
        # Safeguard against zero division
        if weight == 0:
            return None
        
        # Calculate price per ounce
        price_per_ounce = amount / weight
        
        # Calculate cost for 12 ounces and round to 2 decimal places
        cost_per_12_ounces = round(price_per_ounce * 12, 2)
        return cost_per_12_ounces
    
    return None  # Return None if the price string does not match the expected format

# Apply the conversion function to the 'est_price' column
coffee_df['cost_per_12_ounces_usd'] = coffee_df['est_price'].astype(str).apply(convert_to_cost_per_12_ounces)

coffee_df.head()

```

```python
coffee_df.to_csv('coffee_fix_converted.csv', index=False)
```
