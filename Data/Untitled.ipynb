```python
import pandas as pd


# List of all 50 U.S. states
us_states = [
    'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 
    'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 
    'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 
    'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 
    'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio', 
    'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 
    'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 
    'Wisconsin', 'Wyoming'
]

# Define a function to extract the country or state from the location
def extract_country(location):
    if pd.isna(location):
        return None
    parts = location.split(',')
    # If there's only one part, assume it's a country or state
    if len(parts) == 1:
        country_or_state = parts[0].strip()
    else:
        country_or_state = parts[-1].strip()

    # Convert U.S. states to 'US'
    if country_or_state in us_states:
        return 'US'
    return country_or_state

# Apply the function to the location column to create the new location_country column
coffee_df['location_country'] = coffee_df['location'].apply(extract_country)

# Display the updated DataFrame
coffee_df.head()

```

```python
# Extract the country from the 'origin' column and create a new 'origin_country' column
coffee_df['origin_country'] = coffee_df['origin'].apply(lambda x: x.split(',')[-1].strip() if pd.notna(x) else None)

# Display the modified dataframe (optional)
coffee_df.head()
```

```python
coffee_df.to_csv('locations.csv', index=False)
```

```python
location_df = pd.read_csv('locations.csv')
```

```python
# Split the 'origin_country' column where there are multiple countries listed, using semicolon as the delimiter
split_origin_countries = location_df['origin_country'].str.split(';', expand=True)

# Rename the columns to reflect multiple origins
split_origin_countries.columns = [f'origin_country_{i+1}' for i in range(split_origin_countries.shape[1])]

# Combine the split countries with the original data
data_combined = pd.concat([location_df, split_origin_countries], axis=1)

```
